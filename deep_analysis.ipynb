```python
train=df.drop('turn_anti_social',axis=1)
target=df['turn_anti_social']
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
train_std=sc.fit_transform(train)
train.shape
target=df['turn_anti_social']
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import cross_val_score
from sklearn import metrics
```

```python
#we will now different models for checking the accuracy
k_range=list(range(1,31))
k_scores=[]
for i in k_range:
    knn=KNeighborsClassifier(n_neighbors=i)
    scores=cross_val_score(knn,train_std,target,cv=10,scoring='accuracy')
    k_scores.append(scores.mean())
print(k_scores)
    
```

```python
plt.plot(k_range,k_scores)
plt.xlabel("value of k for knn")
plt.ylabel("cross_validated_accuracy")
```

```python
#from the graph it is clear that k=11 give highest score so we use 12 neighbors
knn=KNeighborsClassifier(n_neighbors=11)
print(cross_val_score(knn,train_std,target,cv=10,scoring='accuracy').mean())
```

```python
#lets check other models also
dtc=DecisionTreeClassifier()
print(cross_val_score(dtc,train_std,target,cv=10,scoring='accuracy').mean())
```

```python
#finding best parameters for RandomForestclassifier
rfc_range=list(range(1,20))
rfc_scores=[]
for i in rfc_range:
    rfc=RandomForestClassifier(n_estimators=i)
    scores=cross_val_score(rfc,train_std,target,cv=10,scoring='accuracy')
    rfc_scores.append(scores.mean())
print(rfc_scores)
```

```python
plt.plot(rfc_range,rfc_scores)
plt.xlabel("number of forest for RandomForestClassifier")
plt.ylabel("cross validated accuracy")
```

```python
rfc=RandomForestClassifier(n_estimators=13)
print(cross_val_score(rfc,train_std,target,cv=10,scoring='accuracy').mean())
```
